```python
import pandas as pd

# Leer el archivo CSV
file_path = 'Steam_2024_bestRevenue_1500.csv'
df = pd.read_csv(file_path)

# El siguiente código muestra las primeras 10 filas del archivo
print(df.head(10))
```

```python
# Mostrar información adicional como el tipo de datos en cada columna
print(df.info())

# Opcional: Mostrar estadísticas descriptivas sobre las columnas numéricas
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histograma de precios
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True)
plt.title('Distribución de Precios de los Juegos')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()

# 2. Diagrama de dispersión de reviewScore vs. revenue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='reviewScore', y='revenue', hue='publisherClass')
plt.title('Relación entre el Puntaje de Reseña y los Ingresos')
plt.xlabel('Puntaje de Reseña')
plt.ylabel('Ingresos')
plt.legend(title='Clase del Editor')
plt.show()

# 3. Promedio de tiempo de juego por publisherClass
avg_playtime = df.groupby('publisherClass')['avgPlaytime'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_playtime, x='publisherClass', y='avgPlaytime')
plt.title('Promedio de Tiempo de Juego por Clase del Editor')
plt.xlabel('Clase del Editor')
plt.ylabel('Promedio de Tiempo de Juego (horas)')
plt.show()

# 4. Top 10 juegos por ingresos
top10_revenue = df.nlargest(10, 'revenue')
plt.figure(figsize=(12, 8))
sns.barplot(data=top10_revenue, x='revenue', y='name', palette='viridis')
plt.title('Top 10 Juegos por Ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Nombre del Juego')
plt.show()
```
